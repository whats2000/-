#### 大於 5000 數值抽樣 5000 個

```r
df %>%
  group_by(clarity) %>%
  summarize(n = n()) %>%
  mutate(sample_size = if_else(n > 5000, 5000, n)) %>%
  left_join(df, by = "clarity") %>%
  group_by(clarity) %>%
  summarize(p_value = shapiro.test(sample(price, size = min(n(), sample_size)))$p.value)
```

#### 設置為相同抽樣 500 個

```r
df %>%
  group_by(clarity) %>%
  summarize(p_value = shapiro.test(sample(price, size = 500))$p.value)
```

# 依照原始資料進行

## ANOVA
### 不同族群是否有差異?
* H0: I1 = IF = SI1 = SI2 = VS1 = VS2 = VVS1 = VVS2 = (price)。
* HA: Otherwise

```r
# 將clarity轉為因子變數
df$clarity <- factor(df$clarity)

# 以clarity分群，進行ANOVA檢定
anova_result <- aov(price ~ clarity, data = df)

# 顯示ANOVA分析結果
summary(anova_result)
```

## Follow-up analysis

```r
result <- TukeyHSD(anova_result, conf.level = 0.95)
result
```

```r
plot(result, las = 1, col = "red", cex.axis = 0.5)
```

```r
# 將 TukeyHSD 結果轉換為資料框
tukey_result <- as.data.frame(result$`clarity`)

# 將比較結果依照差異排序
summary_result <- tukey_result[order(tukey_result$diff), ]

# 顯示排序後的比較結果
summary_result
```

## ANOVA
### 不同族群是否有差異?
* H0: I1 = IF = SI1 = SI2 = VS1 = VS2 = VVS1 = VVS2 = (price)。
* HA: Otherwise

```r
# 數據抽樣
df_sampled <- df %>% group_by(clarity) %>% sample_n(size = 500)

# 將clarity轉變成子變因
df_sampled$clarity <- factor(df_sampled$clarity)

# 以clarity分群，進行ANOVA檢定
sample_anova_result <- aov(price ~ clarity, data = df_sampled)

# 顯示ANOVA分析结果
summary(sample_anova_result)
```

## Follow-up analysis

```r
sample_result <- TukeyHSD(sample_anova_result, conf.level = 0.95)
sample_result
```

```r
plot(sample_result, las = 1, col = "red", cex.axis = 0.5)
```

```r
# 將 TukeyHSD 結果轉換為資料框
sample_tukey_result <- as.data.frame(sample_result$`clarity`)

# 將比較結果依照差異排序
sample_summary_result <- sample_tukey_result[order(sample_tukey_result$diff), ]

# 顯示排序後的比較結果
sample_summary_result
```
