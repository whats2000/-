```python
import pandas as pd

df = pd.read_csv('./影響中山大學學生零用金的因素- (回覆)公開版 - 表單回應 1.csv')
df
```

# 針對經濟狀況

```python
# 計算每個家庭經濟狀況描述的出現頻率
economy_counts = df['家庭經濟狀況'].value_counts()
print("家庭經濟狀況頻率:")
economy_counts
```

```python
# 比較不同家庭經濟狀況的受訪者之間的零用金水平
economy_groups = df.groupby('家庭經濟狀況')
average_allowance = economy_groups['目前一個月 的零用金 全部 有多少呢?  \n*單位: 新台幣'].mean()
print("不同家庭經濟狀況的平均零用金:")
average_allowance
```

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='目前一個月 的零用金 全部 有多少呢?  \n*單位: 新台幣', by='家庭經濟狀況')
plt.xlabel('家庭經濟狀況')
plt.ylabel('目前一個月的零用金全部有多少呢? (新台幣)')
plt.title('家庭經濟狀況對目前零用金的影響')
plt.suptitle('')  # 移除預設的子標題
plt.show()
```

```python
from scipy.stats import shapiro

for name, group in df[df['家庭經濟狀況'] != '富裕'].groupby('家庭經濟狀況'):
    stat, p_value = shapiro(group['目前一個月 的零用金 全部 有多少呢?  \n*單位: 新台幣'])
    print(f"家庭經濟狀況: {name}")
    print("Shapiro-Wilk test統計量:", stat)
    print("p值:", p_value)
    print()
```

```python
from scipy.stats import bartlett

# 獨立性檢定 - Bartlett's test
data = [group['目前一個月 的零用金 全部 有多少呢?  \n*單位: 新台幣'] for name, group in df.groupby('家庭經濟狀況')]
stat, p_value = bartlett(*data)
print("Bartlett's test統計量:", stat)
print("p值:", p_value)
```

```python
from scipy.stats import kruskal

# 提取每個群組的零用金資料
data = [group['目前一個月 的零用金 全部 有多少呢?  \n*單位: 新台幣'] for name, group in df.groupby('家庭經濟狀況')]

# 進行Kruskal-Wallis檢定
stat, p_value = kruskal(*data)

# 輸出檢定結果
print("Kruskal-Wallis檢定結果:")
print("統計量:", stat)
print("p值:", p_value)
```

```python
import pandas as pd
from scipy.stats import mannwhitneyu

# 進行兩兩比較的事後分析
groups = df.groupby('家庭經濟狀況')

# 建立空的資料框架
results_df = pd.DataFrame(columns=['Group 1', 'Group 2', 'Mann-Whitney U statistic', 'p-value'])

# 進行多重比較
for name1, group1 in groups:
    for name2, group2 in groups:
        if name1 < name2:  # 避免重複比較
            statistic, p_value = mannwhitneyu(group1['目前一個月 的零用金 全部 有多少呢?  \n*單位: 新台幣'], group2['目前一個月 的零用金 全部 有多少呢?  \n*單位: 新台幣'])

            # 將結果加入資料框架
            results_df = pd.concat([results_df, pd.DataFrame({'Group 1': [name1], 'Group 2': [name2], 'Mann-Whitney U statistic': [statistic], 'p-value': [p_value]})])

# 重新設定索引
results_df = results_df.reset_index(drop=True)

# 顯示資料框架
results_df
```

進行Kruskal-Wallis檢定後，不同組別之間存在統計上的顯著差異，事後分析方法進行多重比較，以比較所有可能的組合。者李使用Mann-Whitney U檢定

```python
import pandas as pd
from scipy.stats import mannwhitneyu

# 進行兩兩比較的事後分析
groups = df.groupby('目前有打工嗎?')

# 建立空的資料框架
results_df = pd.DataFrame(columns=['Group 1', 'Group 2', 'Mann-Whitney U statistic', 'p-value'])

# 進行多重比較
for name1, group1 in groups:
    for name2, group2 in groups:
        if name1 < name2:  # 避免重複比較
            statistic, p_value = mannwhitneyu(group1['目前一個月 的零用金 全部 有多少呢?  \n*單位: 新台幣'], group2['目前一個月 的零用金 全部 有多少呢?  \n*單位: 新台幣'])

            # 將結果加入資料框架
            results_df = pd.concat([results_df, pd.DataFrame({'Group 1': [name1], 'Group 2': [name2], 'Mann-Whitney U statistic': [statistic], 'p-value': [p_value]})])

# 重新設定索引
results_df = results_df.reset_index(drop=True)

# 顯示資料框架
results_df
```

# 探討經濟，打工與家鄉對零用金影響

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder

# 選擇要使用的變數
variables = ['家鄉 (上大學前居住最久的區域)', '目前有打工嗎?', '家庭經濟狀況']

# 提取相關的欄位資料
data = df[variables + ['目前一個月 的零用金 全部 有多少呢?  \n*單位: 新台幣']]

# 檢查缺失值
print(data.isnull().sum())

# 處理缺失值
data = data.dropna()

# 轉換類別變數為獨熱編碼
onehot_encoder = OneHotEncoder(sparse=False)
X_encoded = onehot_encoder.fit_transform(data[variables])
```

```python
# 分割自變數和應變數
X = X_encoded
y = data['目前一個月 的零用金 全部 有多少呢?  \n*單位: 新台幣']

# 建立線性迴歸模型
model = LinearRegression()

# 擬合模型
model.fit(X, y)

# 顯示迴歸係數和截距
print("迴歸係數:")
print(model.coef_)
print("截距:")
print(model.intercept_)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
# 進行預測
y_pred = model.predict(X)

# 計算R平方
r2 = r2_score(y, y_pred)
print("R平方:", r2)

# 計算均方誤差(MSE)
mse = mean_squared_error(y, y_pred)
print("均方誤差(MSE):", mse)
```

```python
# 繪製散點圖和回歸線
plt.scatter(y, y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()], '--', color='red', linewidth=2)
plt.xlabel('實際零用金')
plt.ylabel('預測零用金')
plt.title('實際值 vs. 預測值')
plt.show()
```
